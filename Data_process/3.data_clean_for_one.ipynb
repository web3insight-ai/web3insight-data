```python
import os
import json
import pandas as pd
```

```python
EXTRACTED_DIR = "../Data/extracted/2025/09/01"
CLEANED_DIR = "../Data/cleaned/2025/09/01"
os.makedirs(CLEANED_DIR, exist_ok=True)
```

```python
sample_file = os.path.join(EXTRACTED_DIR, "2025-09-01-1.json")
with open(sample_file, "r") as f:
    lines = f.readlines()
print(f"共 {len(lines)} 条事件")
sample_event = json.loads(lines[0])
sample_event
```

```python
cleaned_events = []
```

```python
with open(sample_file, "r") as f:
    for line in f:
        try:
            event = json.loads(line)

            # 空字段处理
            if event.get("actor", {}).get("login") is None:
                continue

            if event.get("type") == "PushEvent":
                commits = event.get("payload", {}).get("commits", [])
                if not commits:
                    continue

            # 精简字段
            event["actor_login"] = event["actor"]["login"]
            event["repo_name"] = event["repo"]["name"]

            if "gravatar_id" in event["actor"] and not event["actor"]["gravatar_id"]:
                del event["actor"]["gravatar_id"]

            cleaned_events.append(event)

        except Exception as e:
            print(f"❌ 解析失败: {e}")
            continue

# 写入清洗后的结果
output_path = os.path.join(CLEANED_DIR, "2025-09-01-0.cleaned.json")
with open(output_path, "w") as fout:
    for e in cleaned_events:
        fout.write(json.dumps(e) + "\n")

print(f"✅ 清洗完成，输出文件：{output_path}，共保留 {len(cleaned_events)} 条事件。")
```
